# 2.5.1 Principal Component Analysis (PCA)

### 2.5.1.1. Exact PCA and probabilitstic interpretation

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
pca = PCA(n_components=2)  # n_components -> number of components to keep
X_r = pca.fit(X).transform(X)
```

```python
colors = ['navy', 'turquoise', 'darkorange']
```

```python
plt.figure(figsize=(8, 8))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=0.8, linewidth=2, label=target_name)
plt.legend()
```

### 2.5.1.2 Incremental PCA
#### It is typically used as a replacement for principal component analysis (PCA) when the dataset to be decomposed is too large to fit in memory.

```python
import numpy as np
from sklearn.decomposition import IncrementalPCA
```

```python
ipca = IncrementalPCA(n_components=2, batch_size=10)  # batch_size -> number of samples to use for each batch
# used X and y loaded in the previous example
X_ipca = ipca.fit_transform(X)
```

```python
plt.figure(figsize=(8, 8))
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_ipca[y == i, 0], X_ipca[y == i, 1], color=color, label=target_name)
plt.legend()
```
