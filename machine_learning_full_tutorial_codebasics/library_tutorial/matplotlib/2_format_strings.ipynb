# Format Strings in plot Function

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
```

### format strings

```python
# go to https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html  markers for more formating
# instead of assigning color, style to each param, you can use format string to change them
# g -> green; + -> plus marker; -- -> dashed line style
# the order does not matter
plt.plot(x, y, 'g+--')
```

### set transparency

```python
# scale of alpha is 0-1 (not visible - fully visible)
plt.plot(x, y, 'D:b', alpha=0.5)
```
