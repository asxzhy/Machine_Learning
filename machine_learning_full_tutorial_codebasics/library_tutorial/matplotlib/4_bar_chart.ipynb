# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company = ['GOOGLE', 'AMZN', 'MSFT', 'FB']
revenue = [99, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
plt.bar(company, revenue)
```

### xticks()

```python
xpos = np.arange(len(company))

plt.bar(xpos, revenue)
```

```python
# replace inx with company
plt.xticks(xpos, company)
plt.bar(xpos, revenue)
```

### Plot two bars for one company

```python
# the bars are overlapping each other
plt.xticks(xpos, company)
plt.ylabel('revenue(nln)')
plt.title('US Tech Stocks')
plt.bar(xpos, revenue, label='Revenue')
plt.bar(xpos, profit, label='Profit')
plt.legend()
```

#### 

#### Separate two bars

```python
# this is why we are using xticks and np for the bar chart
# the xticks enables us to change the position of different bars
# the np.array() can be subtracted directly and it will generate a new array that is
# been subtracted
plt.xticks(xpos, company)
plt.ylabel('revenue(nln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, label='Revenue')
plt.bar(xpos+0.2, profit, label='Profit')
plt.legend()
```

#### Change the width of bars

```python
plt.xticks(xpos, company)
plt.ylabel('revenue(nln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')
plt.legend()
```

### Horizontal bar chart

```python
# change xticks to yticks inorder to display the company name on the y-axis
plt.yticks(xpos, company)
plt.xlabel('revenue(nln)')
plt.title('US Tech Stocks')

# set the height instead of the width
plt.barh(xpos-0.2, revenue, height=0.4, label='Revenue')
plt.barh(xpos+0.2, profit, height=0.4, label='Profit')
plt.legend()
```
