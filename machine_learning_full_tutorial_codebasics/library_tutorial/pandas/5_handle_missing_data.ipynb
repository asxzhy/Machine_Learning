# Handle Missing Data with fillna, dropna, interpolate

```python
import pandas as pd

df = pd.read_csv('5_weather_data.csv')
df
```

### Convert string to date

```python
type(df.day[0])  # it is read as string
```

```python
# parse the day column to date type
df = pd.read_csv('5_weather_data.csv', parse_dates=['day'])
df
```

```python
df.set_index('day', inplace=True)  # make day column as the index
df
```

### fillna()

```python
new_df = df.fillna(0)  # fill all the NaN value with 0
new_df
```

```python
# by passing in a dictionary with column as key and replace value as value, 
# it replace NaN in columns with different value
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
# forward fill -> fill the NaN value with the previous value
# bfill is backward fill
new_df = df.fillna(method='ffill')
new_df
```

```python
# now, instead of copying the value above the cell, it copys the value to the right
new_df = df.fillna(method='bfill', axis='columns')
new_df
```

```python
# the limit restricts how many time you can carry a value
new_df = df.fillna(method='ffill', limit=1)
new_df
```

### interpolate()

```python
# calculate the NaN value using the value before and after the cell, and equally space out the values
new_df = df.interpolate()
new_df
```

```python
# the temperature for 2017-1-4 should be more close to 2017-1-5 instead of taking the average of 1-1 and 1-5 value
# by specifying the method time, the interpolate now considers the time when calculating NaN values
new_df = df.interpolate(method='time')
new_df
```

### dropna()

```python
# it drops all the rows with at least one NaN value
new_df = df.dropna()
new_df
```

##### how='all'

```python
# only drop the row with all NaN value
new_df = df.dropna(how='all')
new_df
```

##### thresh=int

```python
# thresh value represents at least how many value a row need to have to be remained
new_df = df.dropna(thresh=1)
new_df
```

### Insert missing date

```python
# create a date range, then pass the range to date time index, then reindex your dataframe
dt = pd.date_range('01-01-2017', '01-11-2017')
idx = pd.DatetimeIndex(dt)
df.reindex(idx)
```
