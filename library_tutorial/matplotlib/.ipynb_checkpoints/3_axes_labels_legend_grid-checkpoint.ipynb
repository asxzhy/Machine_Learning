# Axes labels, Legend, Grid

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

### plot multiple lines in one plot

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

### Axes Labels

```python
plt.xlabel('Days')
plt.xlabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

### Set Legend

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

plt.xlabel('Days')
plt.xlabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

# need to call this function in order to see the legend
plt.legend()
```

#### Property of legend()

```python
plt.xlabel('Days')
plt.xlabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

# by default, loc is set to 'best' which find the best location for displaying a legend
# shadow makes legend look a bit nicer
plt.legend(loc='upper right', shadow=True, fontsize='large')
```

### grid( )

```python
plt.xlabel('Days')
plt.xlabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

plt.legend()

# creates grid on the plot
plt.grid()
```
