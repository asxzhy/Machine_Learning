# Introduction to matplotlib

```python
import matplotlib.pyplot as plt

# the following code is used to display the graph in the notebook instead of poping up a new window
# it is not used here because the backend is already set to inline
# you can check the backend with matplotlib.get_backend()
# %matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]  # weekdays
y = [50, 51, 52, 48, 47, 49, 46]  # temperature
```

### plt.plot( )

```python
# to plot the x y values
plt.plot(x, y)
```

```python
# the different properties of the plot can be changed
plt.plot(x, y, color='green', linewidth=5, linestyle='dashdot')
```

#### Adding label to x, y axis and the title of the plot

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x, y, color='green')
```
