# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food', 'Phone/Interner Bill', 'Car', 'Other Utilities']
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

### 

### pie( ) properties
#### pie size

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5)
```

#### show percentage

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%')
```

#### make a part of the pie to come out

```python
# by setting a value to explode, the part will come out based on the value you give it
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0, 0.2, 0, 0, 0])
```

#### pie position

```python
# set the startangel to set the position of the pie
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True, explode=[0, 0.2, 0, 0, 0],
       startangle=90)
```
